## 01 - Persiapan dan Konteks Data

**Tujuan Notebook:** Menyiapkan dataset untuk analisis, memberikan konteks tentang asal-usul data, dan memisahkannya menjadi set data untuk training model dan simulasi live.

### Konteks Proyek

Lingkungan spektrum frekuensi radio (RF) adalah sumber daya yang terbatas dan semakin padat. Kemampuan untuk memonitor spektrum secara otomatis dan mendeteksi aktivitas yang tidak biasa (anomali) sangat penting untuk berbagai aplikasi, mulai dari manajemen spektrum, keamanan nirkabel, hingga identifikasi interferensi.

Proyek ini bertujuan untuk membangun sebuah sistem *unsupervised* yang dapat:
1. Menganalisa data spektrum mentah.
2. Mengidentifikasi dan mengelompokkan pola-pola sinyal yang ada.
3. Mendeteksi dan memberikan skor pada sinyal anomali secara real-time.

**Metodologi Data:**
*   **Sumber Data:** Pengukuran dilakukan menggunakan *Software-Defined Radio* (SDR) **HackRF One**.
*   **Rentang Frekuensi:** Pengukuran difokuskan pada rentang **700 MHz hingga 1900 MHz**, mencakup pita frekuensi seluler, navigasi, dan penyiaran yang vital.
*   **Simulasi Anomali:** Untuk melatih dan menguji sistem, sinyal anomali yang terkontrol disuntikkan ke lingkungan RF menggunakan **Signal Generator**. Sinyal-sinyal ini sengaja ditempatkan pada frekuensi atau dengan karakteristik yang tidak biasa untuk menantang kemampuan deteksi sistem.

### Inisialisasi & Pemuatan Data

```python
import os
import glob

# Definisikan path ke data mentah
DATA_PATH = '../data/raw/' # Path relatif dari folder notebooks

# Ambil semua file pengukuran
all_files = sorted(glob.glob(os.path.join(DATA_PATH, '*')))

print(f"Total file pengukuran ditemukan: {len(all_files)}")
```

### Pemisahan Data: Training vs. Live Simulation

Untuk mensimulasikan skenario dunia nyata, kita akan memisahkan data kita. Sebagian besar data akan digunakan untuk melatih model kita dalam mengenali 'kondisi normal' di spektrum. Beberapa file akan kita sisihkan untuk bertindak sebagai data 'live' atau 'tidak dikenal' yang akan dianalisa oleh sistem yang sudah terlatih.

```python
# Pilih beberapa file secara manual untuk simulasi live
# File-file ini akan dikecualikan dari proses training
live_files_paths = [
    os.path.join(DATA_PATH, '1700_anomali_1'),
    os.path.join(DATA_PATH, '1000_anomali_4')
]

# Buat daftar file training dengan mengecualikan file live
training_files_paths = [f for f in all_files if f not in live_files_paths]

print(f"Jumlah file untuk training model: {len(training_files_paths)}")
print(f"Jumlah file untuk simulasi live: {len(live_files_paths)}")
```
